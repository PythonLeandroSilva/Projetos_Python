```python

#Neste Case de Call Center estou trabalhando usando uma planilha com dados
#Onde usarei o pandas para analisar indicadores

# Importe a biblioteca Pandas
import pandas as pd

# Carregue o arquivo Excel em um DataFrame
agentes_df = '/content/drive/MyDrive/dados/agentes.xls'
df = pd.read_excel(agentes_df)
display(df)
```

```python
#PEGAR LINHAS QUE CORRESPONDEM A UMA CONDIÇÃO
SETOR1A = (df.loc[df['SETOR'] == '1A'])
display(SETOR1A)
```

```python
# Aqui foi criado função para converter o total que está em segundos nas colunas
# TMA e ACW em formato de relógio (hh:mm:ss)

def segundos_para_relogio(segundos):
    try:
        segundos = int(segundos)  # Converter para inteiro
        horas = segundos // 3600
        minutos = (segundos % 3600) // 60
        segundos = segundos % 60
        return f"{int(horas):02d}:{int(minutos):02d}:{int(segundos):02d}"
    except ValueError:
        return segundos  # Se não for possível converter, retorne o valor original

# Aplicar a função à coluna "TMA"
df["TMA"] = df["TMA"].apply(segundos_para_relogio)

# Aplicar a função à coluna "ACW"
df["ACW"] = df["ACW"].apply(segundos_para_relogio)

display(df)

```
