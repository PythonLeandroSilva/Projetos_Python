```python
import pandas as pd

# Carregue o arquivo Excel em um DataFrame
df = '/content/sample_data/dados/base.xlsx'
dados = pd.read_excel(df)

# Arredonde as colunas para duas casas decimais
colunas_para_arredondar = ["TME REAL", "TME PREVISTO"]
dados[colunas_para_arredondar] = dados[colunas_para_arredondar].round(2)

# Formate a coluna "OPERADORES PREVISTOS" sem casas decimais
dados["OPERADORES PREVISTOS"] = dados["OPERADORES PREVISTOS"].round(0).astype(int)
dados["OPERADORES PRESENTES"] = dados["OPERADORES PRESENTES"].round(0).astype(int)
dados["PREVISÃO DE LIGAÇÕES"] = dados["PREVISÃO DE LIGAÇÕES"].round(0).astype(int)

# Formate a coluna "PCA" como porcentagem com o símbolo "%"
dados["PCA"] = (dados["PCA"] * 100).round(0).astype(int).astype(str) + "%"

# Função para converter segundos em formato de relógio (hh:mm:ss)
def segundos_para_relogio(segundos):
    horas = segundos // 3600
    minutos = (segundos % 3600) // 60
    segundos = segundos % 60
    return f"{int(horas):02d}:{int(minutos):02d}:{int(segundos):02d}"

# Aplicar a função à coluna "TME REAL"
dados["TME REAL"] = dados["TME REAL"].apply(segundos_para_relogio)

# Aplicar a função à coluna "TME PREVISTO"
dados["TME PREVISTO"] = dados["TME PREVISTO"].apply(segundos_para_relogio)

# Calcular a média da coluna "TME REAL"
media_tme_real = dados["TME REAL"].str.split(':').apply(lambda x: int(x[0]) * 3600 + int(x[1]) * 60 + int(x[2])).mean()

# Converter a média de segundos para o formato de relógio (hh:mm:ss)
media_tme_real_formatado = segundos_para_relogio(media_tme_real)

# Criar a coluna "MÉDIA TME" e atribuir o valor formatado
dados["MÉDIA TME"] = media_tme_real_formatado

# Calcular a média da coluna "TME PREVISTO"
media_tme_previsto = dados["TME PREVISTO"].str.split(':').apply(lambda x: int(x[0]) * 3600 + int(x[1]) * 60 + int(x[2])).mean()

# Converter a média de segundos para o formato de relógio (hh:mm:ss)
media_tme_previsto_formatado = segundos_para_relogio(media_tme_previsto)

# Criar a coluna "MÉDIA TME PREVISTO" e atribuir o valor formatado
dados["MÉDIA TME PREVISTO"] = media_tme_previsto_formatado

# Exibir o DataFrame com as colunas formatadas
display(dados)
```
